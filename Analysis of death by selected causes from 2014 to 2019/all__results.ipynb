<b>Data Description</b>:
1.	<b>Year</b>: Represents the year in which the data was recorded.
2.	<b>Month</b>: Indicates the specific month within the year when the data was collected.
3.	<b>All Cause</b>: The total number of deaths due to all causes in a given year and month.
4.	<b>Natural Cause</b>: The number of deaths attributed to natural causes, which typically includes climatic conditions or disasters.
5.	<b>Septicemia</b>: The number of deaths caused by septicemia, a severe bloodstream infection.
6.	<b>Malignant Neoplasms</b>: Represents deaths caused by malignant neoplasms, which are cancerous tumors.
7.	<b>Diabetes Mellitus</b>: The number of deaths resulting from diabetes mellitus, a metabolic disorder characterized by high blood sugar levels.
8.	<b>Alzheimer Disease</b>: Indicates deaths related to Alzheimer's disease, a neurodegenerative condition that affects memory and cognitive function.
9.	<b>Influenza and Pneumonia</b>: Represents deaths due to influenza (flu) and pneumonia, typically respiratory infections.
10.	<b>Chronic Lower Respiratory Diseases</b>: The number of deaths resulting from chronic lower respiratory diseases, which include conditions like chronic obstructive pulmonary disease (COPD).
11.	<b>Other Diseases of Respiratory System</b>: Deaths attributed to other diseases of the respiratory system.
12.	<b>Nephritis, Nephrotic Syndrome, and Nephrosis</b>: Indicates deaths related to kidney diseases, including nephritis and nephrotic syndrome.
13.	<b>Symptoms, Signs, and Abnormal Clinical and Laboratory Findings, Not Elsewhere Classified</b>: Represents deaths where the primary cause was not a specific disease but rather abnormal clinical or laboratory findings.
14.	<b>Diseases of Heart</b>: The number of deaths caused by heart diseases, including conditions like coronary artery disease and heart failure.
15.	<b>Cerebrovascular Diseases</b>: Indicates deaths due to cerebrovascular diseases, such as strokes.
16.	<b>Accidents (Unintentional Injuries)</b>: Represents deaths resulting from unintentional accidents.
17.	<b>Motor Vehicle Accidents</b>: The number of deaths specifically related to motor vehicle accidents.
18.	<b>Intentional Self-Harm (Suicide)</b>: Indicates deaths due to suicide, where individuals intentionally take their own lives.
19.	<b>Assault (Homicide)</b>: Represents deaths due to assaults or homicides, where one person intentionally causes the death of another.
20.	<b>Drug Overdose</b>: Indicates deaths resulting from drug overdoses, often associated with the abuse or misuse of drugs or substances.
