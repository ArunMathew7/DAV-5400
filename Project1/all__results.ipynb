<h5><b>Analyzing which what is the cause in leading deaths.</b></h5>

These plots illustrate the leading causes of death and their respective total death counts, making it easy to identify the primary causes of mortality in the dataset.

<h5><b>Analyzing trends in deaths due to external factors</b></h5>

The resulting plot displays the trends over time for accidents (motor vehicle and unintentional), suicides or homicides and drug overdoses.

```python
EDA.EDA_analysis.external_factors()
```

<h5><b>As “Natural Cause” is the leading cause of death analyzing the count of death for each month for the years 2014 to 2019.</b></h5>

As we can see through the analysis for every year the number of deaths is high for the month of January for natural causes,
